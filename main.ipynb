```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

# Guilherme de Abreu 22.222.028-7
# Matheus Cucio 22.121.035-4

# Carregar o arquivo CSV
df = pd.read_csv("winequality-red.csv")

# Verificar valores duplicados
num_duplicated_rows = df.duplicated().sum()
print("Número de linhas duplicadas:", num_duplicated_rows)

# Remover linhas duplicadas
df.drop_duplicates(inplace=True)

# Definir o target (y) e o atributo (x)
x = df['alcohol']
y = df['quality']

# a) Média, variância, desvio padrão e mediana para x e y
print("\na)\n")
print("\nEstatísticas para x (alcohol):")
print("   - Média:", x.mean())
print("   - Variância:", x.var())
print("   - Desvio padrão:", x.std())
print("   - Mediana:", x.median())

print("\nEstatísticas para y (quality):")
print("   - Média:", y.mean())
print("   - Variância:", y.var())
print("   - Desvio padrão:", y.std())
print("   - Mediana:", y.median())


# b) Histograma de x e y
print("\nb)\n")
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(x, bins=20, color='blue', alpha=0.7)
plt.title('Histograma de x')
plt.xlabel('Valores de x')
plt.ylabel('Frequência')

plt.subplot(1, 2, 2)
plt.hist(y, bins=20, color='red', alpha=0.7)
plt.title('Histograma de y')
plt.xlabel('Valores de y')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()

# c) Boxplot de x e y
print("\nc)\n")
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(x=x, color='blue')
plt.title('Boxplot de x')

plt.subplot(1, 2, 2)
sns.boxplot(x=y, color='red')
plt.title('Boxplot de y')

plt.tight_layout()
plt.show()

# d) Coeficiente de correlação de x e y
print("\nd)\n")
correlation = df.corr().loc['alcohol', 'quality']
print("\nCoeficiente de correlação entre x (alcohol) e y (quality):", correlation)

# e) Teste de normalidade para x e y
print("\ne)\n")
stat_x, p_x = shapiro(x)
stat_y, p_y = shapiro(y)
print("\nTeste de normalidade para x (alcohol):")
print("   - Estatística:", stat_x)
print("   - p-valor:", p_x)

print("\nTeste de normalidade para y (quality):")
print("   - Estatística:", stat_y)
print("   - p-valor:", p_y)

# f) Gráfico de densidade junto com histograma para as variáveis x e y
print("\nf)\n")
plt.figure(figsize=(10, 6))
sns.histplot(x, kde=True, color='blue', label='x', stat='density')
sns.histplot(y, kde=True, color='red', label='y', stat='density')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.title('Gráfico de densidade')
plt.legend()
plt.show()
```
